```python
from tkinter import *
```

```python
root=Tk()
root.title("calculator")

e=Entry(root,width=41,borderwidth=5)
e.grid(column=0,row=0,columnspan=3,pady=15,)

def click(num):
    num=str(num)
    current=e.get()
    curren=str(current)
    lst=["+","-","x","/"," "]
    if num in lst and current[-1] in lst:
      return
    else:
      e.delete(0,END)
      e.insert(0,str(current)+str(num))

def clean1():
   current=e.get()
   e.delete(0,END)
   current=current[:-1]
   e.insert(0,current)
    
def cleanall():
   e.delete(0,END)
   
def equate():
   s=e.get()
   s=" "+s+" "
   list=["/","x","+","-"," "]
   while(s.count("/")!=0):
      a=s.index("/")
      i=a-1
      while(True):
            
            if(s[i] in list):
              break
            i=i-1
      j=a+1
      while(True):
         
         if(s[j] in list):
            break
         
         j=j+1
     
      x=int(s[i+1:a])
      y=int(s[a+1:j])
      z=x/y
      s=s[:i+1]+str(z)+s[j:]

   while(s.count("x")!=0):
      a=s.index("x")
      i=a-1
      while(True):
            
            if(s[i] in list):
              break
            i=i-1
      j=a+1
      while(True):
         
         if(s[j] in list):
            break
         
         j=j+1
     
      x=int(s[i+1:a])
      y=int(s[a+1:j])
      z=x*y
      s=s[:i+1]+str(z)+s[j:]
     
   while(s.count("+")!=0):
      a=s.index("+")
      i=a-1
      while(True):
            
            if(s[i] in list):
              break
            i=i-1
      j=a+1
      while(True):
         
         if(s[j] in list):
            break
         
         j=j+1
     
      x=int(s[i+1:a])
      y=int(s[a+1:j])
      z=x+y
      s=s[:i+1]+str(z)+s[j:]
   while(s.count("-")!=0):
      a=s.index("-")
      i=a-1
      while(True):
         
            if(s[i] in list):
              break
            i=i-1
      j=a+1
      while(True):
         
         if(s[j] in list):
            break
         
         j=j+1
     
      x=int(s[i+1:a])
      y=int(s[a+1:j])
      z=x-y
      s=s[:i+1]+str(z)+s[j:]

   e.delete(0,END)
   e.insert(0,s)


but1=Button(root,text="1",padx=30,pady=30,command=lambda: click(1))
but2=Button(root,text="2",padx=30,pady=30,command=lambda: click(2))
but3=Button(root,text="3",padx=30,pady=30,command=lambda: click(3))
but4=Button(root,text="4",padx=30,pady=30,command=lambda: click(4))
but5=Button(root,text="5",padx=30,pady=30,command=lambda: click(5))
but6=Button(root,text="6",padx=30,pady=30,command=lambda: click(6))
but7=Button(root,text="7",padx=30,pady=30,command=lambda: click(7))
but8=Button(root,text="8",padx=30,pady=30,command=lambda: click(8))
but9=Button(root,text="9",padx=30,pady=30,command=lambda: click(9))
but0=Button(root,text="0",padx=30,pady=30,command=lambda: click(0))
but_add=Button(root,text="+",padx=30,pady=30,command=lambda: click("+"))
but_sub=Button(root,text="-",padx=30,pady=30,command=lambda: click("-"))
but_mul=Button(root,text="x",padx=30,pady=30,command=lambda: click("x"))
but_div=Button(root,text="/",padx=30,pady=30,command=lambda: click("/"))
but_equal=Button(root,text="=",padx=30,pady=30,command=equate)
but_clear=Button(root,text="clear",padx=30,pady=30,command=cleanall)
but_del=Button(root,text="del",padx=30,pady=30,command=clean1)
but1.grid(row=6,column=0)
but2.grid(row=6,column=1)
but3.grid(row=6,column=2)

but4.grid(row=5,column=0)
but5.grid(row=5,column=1)
but6.grid(row=5,column=2)

but7.grid(row=4,column=0)
but8.grid(row=4,column=1)
but9.grid(row=4,column=2)

but_add.grid(row=4,column=4)
but_sub.grid(row=5,column=4)
but_mul.grid(row=6,column=4)
but_div.grid(row=7,column=4)

but0.grid(column=1,row=7)
but_clear.grid(column=0,row=7)
but_del.grid(column=2,row=7)

but_equal.grid(row=8,column=0)

root.mainloop()
```
